```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# Setup driver
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

url = "https://www.imdb.com/search/title/?title_type=feature&release_date=2024-01-01,2024-12-31"
driver.get(url)
time.sleep(3)

movies = []
storylines = []

# Each movie block
containers = driver.find_elements(By.XPATH, "//li[contains(@class,'ipc-metadata-list-summary-item')]")

for movie in containers:
    # --- Movie Title ---
    try:
        title = movie.find_element(By.XPATH, ".//h3[contains(@class, 'ipc-title__text')]").text
        title = title.split(". ", 1)[-1].strip()   # remove "1. " etc.
    except:
        title = "N/A"
    
    # --- Storyline ---
    try:
        storyline = movie.find_element(By.XPATH, ".//div[contains(@class, 'ipc-html-content')]").text
    except:
        storyline = "N/A"
    
    movies.append(title)
    storylines.append(storyline)

driver.quit()

# Save to DataFrame
df = pd.DataFrame({"Movie Name": movies, "Storyline": storylines})
df.to_csv("imdb_movies_2024.csv", index=False, encoding="utf-8-sig")
df.head()
```
